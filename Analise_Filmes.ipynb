Usando factorize - Essa medida tambÃ©m inclui os valores categoricos

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
plt.figure(figsize=(10, 8), dpi=80)
sns.heatmap(correlation_matrix, annot = True)

plt.title("Matriz de correlaÃ§Ã£o dos filmes")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

Visualizanado melhor utilizando listas

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

Verificando correlaÃ§Ãµes maiores que 0.5

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

Top 15 empresas com valor bruto

```python
ValorTotal = df.groupby('company')[["gross"]].sum()

ValorTotalSorted = ValorTotal.sort_values('gross', ascending = False)[:15]

ValorTotalSorted = ValorTotalSorted['gross'].astype('int64') 

ValorTotalSorted
```

```python
ValorTotalSum = df.groupby(['company', 'year'])[["gross"]].sum()

ValorTotalSumSorted = ValorTotalSum.sort_values(['gross','company','year'], ascending = False)[:15]

ValorTotalSumSorted = ValorTotalSumSorted['gross'].astype('int64') 

ValorTotalSumSorted
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
plt.figure(figsize=(10, 8), dpi=80)
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
